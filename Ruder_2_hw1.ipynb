**(a)** Following the example in the lab, reproduce the following density plot for petal length with each iris species. What insights can you derive by looking at this density plot?: _1 points_

p2a.png

````r
# YOUR SOLUTION HERE
ggplot(data = iris, mapping = aes(x = Petal.Length, color = Species, fill = Species))+
  geom_density(alpha = 0.5) +
  labs(title = 'Density plot of petal length across species', x = 'Petal length (cm)', y = 'Species')
# Insights: The density plot suggests the Setosa species generally has the shortest petal length among the three species of iris.
# The petal lengths of the Versicolor and Virginica species seem to overlap, suggesting a stronger genetic relation. The data also indicates
# greater genetic variation among the latter two species, as the petal lenghth measurements seem to spread more.
````

**(b)** How are petal length and petal width related? Following the example we saw in lecture, reproduce the following plot: _1 points_


p2b.png

````r
# YOUR SOLUTION HERE
# Petal length and petal width seem to be positively correlated: As petal length increases, petal width increases.
ggplot(data = iris, mapping = aes(x = Petal.Length, y = Petal.Width))+
  geom_point(aes(color = Species, shape = Species)) +
  geom_smooth() +
  labs(title = 'Petal length and width for iris flowers', x = 'Petal length (cm)', y = 'Petal width (cm)')
````

**(c)**
The preceding plot indicates that the petal width is roughly linearly related to the petal length. Check this by computing the correlation of petal width  with petal length, and round the correlation to two decimal places. (*Hint*: R contains a built-in function; for computing the correlation of two vectors try `help(cor)` to learn more; and similarly another built-in function for rounding values to a specific number of decimal places.) _1 point_

````r
# YOUR SOLUTION HERE
cor(iris$Petal.Length, iris$Petal.Width, use = "everything", method = ("pearson"))
# rounded correlation: 0.96
# 0.96 indicates a strong positive correlation between petal length and petal width
````
